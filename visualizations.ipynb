# Energy loss for energy-dependent decay rates


```python
%matplotlib inline
from energyloss import EnergyLoss as EL
el_ct = EL (constant_loss=0.8, glashow_unit=0.8)
el_var = EL (constant_loss=None, glashow_unit=0.8)
```

```python
%%capture
el_ct.final_energy();
```



### In the two next graphs, we can see the great agreement between the numerical simulation within energy_loss.py and the analytical approach in http://inspirehep.net/record/1276809 (see figures 2 and 3). In addition, we can see the particles that end with more than 1.5 times the terminal energy


```python
el_ct.final_initial(std=True, n_decays=True);
```

```python
%%capture
el_ct.propagation(distance=1e+4);
```

```python
el_ct.final_propagation(std=True);
```



### In the next figure, we remove the approximation that the energy loss per decay is constant and take into account that the energy loss is a random number given a probability distribution. The most significant differences are two: first, more particles end up with energies higher than 1.5 times the terminal energy. Second, the fluctuating character of the energy loss per decay breaks the regularities in the graph above.
